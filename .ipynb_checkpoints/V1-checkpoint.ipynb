Pipeline

```python
# Data cleaning

#cleaning

def clean_data(df):
    df.columns=[e.lower().replace(' ', '_') for e in df.columns] #all columns in lower case and with _
    df =df.drop_duplicates()
    return df

hs_df_numerical = hs_df.select_dtypes(include=np.number)
hs_df_categorical = hs_df.select_dtypes(include=np.object)
```

```python
# set a grey background (use sns.set_theme() if seaborn version 0.11.0 or above) 
sns.set(style="darkgrid")
df = hs_df
 
# plotting both distibutions on the same figure
fig = sns.kdeplot(df['number_of_times_contacted'], shade=True, color="r")
fig = sns.kdeplot(df['number_of_sales_activities'], shade=True, color="b")
fig.legend(labels=['number_of_times_contacted','number_of_sales_activities'])
fig.set(xlabel='number_of_contacts', ylabel='density')

plt.show()
```

```python
# Create bars
barWidth = 0.9
bars1 = hs_df.number_of_times_contacted
bars2 = hs_df.number_of_sales_activities

 
# The X position of bars
r1 = [1,3,5]
r2 = [2,4,6]

 
# Create barplot
plt.bar(r1, bars1, width = barWidth, color = (0.3,0.1,0.4,0.6), label='number_of_times_contacted')
plt.bar(r2, bars2, width = barWidth, color = (0.3,0.5,0.4,0.6), label='number_of_sales_activities')

 
# Create legend
plt.legend()
 
# Text below each barplot with a rotation at 90°
#plt.xticks(hs_df.deal_stage, rotation=90)
 

# Adjust the margins
#plt.subplots_adjust(bottom= 0.2, top = 0.98)
 
# Show graphic
plt.show()
```

```python
# set plot style: grey grid in the background:
sns.set(style="darkgrid")


# Set the figure size
plt.figure(figsize=(8, 8))
plt.xticks(rotation=70)
plt.tight_layout()

N = 2
ind = np.arange(N)  # the x locations for the groups
width = 0.27  

x= hs_df.deal_stage

# grouped barplot
sns.barplot(x, y="number_of_sales_activities", data=hs_df, ci=None)
sns.barplot(x, y="number_of_times_contacted",data=hs_df, ci=None)
```
