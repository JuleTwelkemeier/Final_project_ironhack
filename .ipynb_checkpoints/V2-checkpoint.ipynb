Pipeline

```python
## Hubspot dataset into df's
hs_df = pd.read_csv('data/hubspot_data_01.10.21.csv')


## Data cleaning

#cleaning

def clean_data(df):
    df.columns=[e.lower().replace(' ', '_') for e in df.columns] #all columns in lower case and with _
    #df =df.drop_duplicates() #no duplicates in HS dataset_
    #df = df.dropna(how ='all') # delete rows which are completly empty :not existing HS_dataset
    df['close_date']=pd.to_datetime(df['close_date'])
    df['last_activity_date'] = pd.to_datetime(df['last_activity_date'])
    df['last_contacted'] = pd.to_datetime(df['last_contacted'])
    df['last_modified_date'] = pd.to_datetime(df['last_modified_date'])
    df['decision_timeline'] = pd.to_datetime(df['decision_timeline'])
    df['estimated_go_live'] = pd.to_datetime(df['estimated_go_live'])
    return df.head()

clean_data(hs_df)
```

Checking Multicorrelations

```python
# creating dummies for hs_df_categorical

hs_df_dummies = pd.get_dummies(data=hs_df, columns =hs_df_categorical,drop_first =True)
hs_df_dummies.sample(10)
```

```python
correlations = hs_df_dummies.corr()
print(correlations)
cmap = sns.cm.rocket_r  ## for reversing the colors
mask=np.zeros_like(hs_df_dummies.corr())
mask[np.triu_indices_from(mask)]=True
fig, ax=plt.subplots(figsize=(10, 8))
ax=sns.heatmap(hs_df_dummies.corr(), mask=mask, annot=True,cmap=cmap)
plt.show()
```
